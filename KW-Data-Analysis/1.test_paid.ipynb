```python
# 檢查數據
import pandas as pd
data = pd.ExcelFile(r"C:\Users\user\Desktop\Python\Martech\Python關鍵字分析\new_folder\1_ cathay-ins(2024_08_05-2024_09_01) .xlsx")
df = data.parse()
df.info()
df.head(5)        
```

```python
# 只採用以下三個欄位

df = df[['关键词', '点击量', "热门网址"]]
df.head(5)

# 檢查
str_rows = df['点击量'].astype(str).sum()
print(str_rows)
```

```python
# 將"点击量"不是int的部分拿出來看
non_int_rows = df[df['点击量'].apply(lambda x: not isinstance(x, int))] # df[col'].apply(lambda x: isinstance(x, str)) 會返回一個與 df['col'] 具有相同長度的布林序列
click_counts = non_int_rows['点击量']
a = len(click_counts)
print(a)
click_counts 
```

```python
unique_value = non_int_rows['点击量'].unique()
print(unique_value)
```

```python
# 由於影響不大所以直接轉換成0
df['点击量'] = df['点击量'].apply(lambda x: 0 if isinstance(x, str) else x)
df['点击量'].sum()
df['点击量'].info()
```

## 分組

```python
# 有"國泰"的關鍵字 # str.contains()
filtered_df = df[df['关键词'].str.contains('國泰', na=False)] # df[布林條件]：篩選出符合的行。
filtered_df.info() # 102 個
filtered_df.head(10) 
```

```python
# 無"國泰"
# "~": 對布林值進行取反操作的運算符
filtered_out_df = df[~df['关键词'].str.contains('國泰', na=False)]
filtered_out_df.info()
filtered_out_df
```

```python
filtered_clean_df = filtered_out_df.dropna() # dropna 也有 inplace 參數
filtered_clean_df.info()
r = filtered_clean_df["关键词"].astype(str).sum()
print(r)
filtered_clean_df
```

```python
# 無"國泰"關鍵字
# 按 '網址' 分組，聚合點擊數，並將關鍵字收集為列表
grouped_df1 = filtered_clean_df.groupby('热门网址', as_index=False).agg({
    '点击量': 'sum',        
    '关键词': lambda x: ', '.join(x)  # 將關鍵字聚合為一個逗號分隔的字串
})

# 對聚合結果按點擊數進行降序排序
sorted_grouped_df1 = grouped_df1.sort_values(by='点击量', ascending=False)
sorted_grouped_df1.head(5)
```

```python
# 有"國泰"關鍵字
grouped_df2 = filtered_df.groupby('热门网址', as_index=False).agg({
    '点击量': 'sum',        
    '关键词': lambda x: ', '.join(x)  # 將關鍵字聚合為一個逗號分隔的字串
})

# 對聚合結果按點擊數進行降序排序
sorted_grouped_df2 = grouped_df2.sort_values(by='点击量', ascending=False)
sorted_grouped_df2
```

## set

```python
# 關鍵字改成set
sorted_grouped_df1['关键词'] = sorted_grouped_df1['关键词'].apply(lambda x: set(x.split(', ')))
```

```python
sorted_grouped_df2['关键词'] = sorted_grouped_df2['关键词'].apply(lambda x: set(x.split(', ')))
```

## rename columns

```python
sorted_grouped_df1 = sorted_grouped_df1.rename(columns={'关键词': 'Keyword', '点击量': 'Clicks', '热门网址': 'Website'})
sorted_grouped_df1.head(5)
```

```python
sorted_grouped_df2 = sorted_grouped_df2.rename(columns={'关键词': 'Keyword', '点击量': 'Clicks', '热门网址': 'Website'})
sorted_grouped_df2.head(5)
```
