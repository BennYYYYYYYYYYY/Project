# 計算任務 TOKEN 數


```python
import pandas as pd
import os
import tiktoken 
import textwrap
```

```python
# 將 df sentence 插入 prompt 中
def sentence_injection(value):
    content = {
    "Sentiment":textwrap.dedent(f'''

    '''),

    "ESG":textwrap.dedent(f'''

    ''')
    }

    return content

def definition_prompt_new(value):
    content = {
    "Readability":textwrap.dedent(f'''

    '''),


    "Subjectivity":textwrap.dedent(f'''

    '''),


    "GreenWashing":textwrap.dedent(f'''

    '''),


    "Ambiguity":textwrap.dedent(f'''

    '''),


    "Timeliness":textwrap.dedent(f'''

    '''),


    "Sentiment":textwrap.dedent(f'''

    '''),


    "ESG":textwrap.dedent(f'''

    ''')
    }
    
    return content

def token_count(folder, prompt_topic):
    encoding = tiktoken.encoding_for_model("gpt-4o")
    total_token_num = 0
    total_counter = 0
    for file in os.listdir(folder):
        file_xlsx = os.path.join(folder, file)
        
        excel_file_token = 0
        row_counter = 0

        df = pd.read_excel(io=file_xlsx, usecols=["sentence"])
        for value in (df["sentence"]):
        
            row_token = 0

            content = definition_prompt_new(value) # 注入 df 中每一 row 的句子進入設計 prompt 中

            messages = [
                {"role":"system", "content":''}, 
                {"role":"user", "content":content[prompt_topic]}
            ]

            for message in messages:
                row_token += len(encoding.encode(message["role"]))
                row_token += len(encoding.encode(message["content"]))
            
            row_counter += 1
            # print(f"這 row 有 {row_token} 個 TOKEN 數")
            excel_file_token += row_token

        total_counter += row_counter
        total_token_num += excel_file_token
        # print(f"第{total_counter}個 Excel 檔案的總 token 數為 {excel_file_token}，並累積到 {total_token_num} 個 token 了")
        

    print(f"共有 {total_counter} 列數據")
    print(f"===================== 全部關於 '{prompt_topic}' 的 token 數為 {total_token_num} ==================================")
```

```python
import time
folder = r"sentence_data"

topics = ['Readability', 'Subjectivity', 'GreenWashing', 'Ambiguity', 'Timeliness', 'Sentiment', 'ESG']

for topic in topics:
    token_count(folder, topic)
    time.sleep(1)    

# token_count(folder, "Sentiment")
# token_count(folder, "ESG")
```
